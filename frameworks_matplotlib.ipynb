```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=np.random.randn(1000)
fig=plt.figure(figsize=(5,3)) 
plt.hist(data, 
         color='skyblue',
         edgecolor='black',
         bins=30)
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
genre=['Drama','Sci-Fi','Romance','Comedy','Action'] 
data=[40,20,10,5,30] 
color=['skyblue', 'lightgreen', 'lightyellow', 'lightgrey', 'pink']
fig=plt.figure(figsize=(5,3)) 
plt.pie(data,labels=genre,colors=color) 
plt.show
```

```python
Country=('China', 'India', 'United States', 'Indonesia', 'Brazil', 'Pakistan', 'Bangladesh', 'Nigeria')
Population=(1.3,1.1, 0.7,0.5,0.3,0.2,0.2,0.1) 
plt.figure(figsize=(8,4))
plt.bar(Country, Population,color="skyblue", width=0.5)
plt.xlabel('COUNTRIES')
plt.ylabel('POPULATION IN BILLIONS') 
plt.title('HOST POPULATED COUNTRIES')
plt.show()
```

```python
a=np.array([1,2,3])
y=np.min(a,axis=None,out=None)
print(y)

b= np.array([1, 2, 3, 4, 5])
z = np.mean(b, axis=None)
print(z)

c = np.array([1, 2, 3, 4, 5])
w = np.max(c, axis=None,out=None)
print(w)

d= np.array([1, 2, 3, 4, 5])
v=np.unique(d,return_index=False,return_inverse=False,return_counts=False)
print(v)
e = np.array([1, 2, 3, 4, 5])
u = np.median(e, axis=None)
print(u)
```

```python
a = np.array([1, 2, 3, 4, 5])
bins=[1,2]
x = np.digitize(a, bins,right=False)
print(x)

b = np.array([1, 2, 3, 4, 5, 6])
y = np.reshape(b, newshape=(2, 3))
print(y)

c = np.array([1, 0, 2, 0, 3])
z= np.count_nonzero(c, axis=None, keepdims=False)
print(z)
d = np.array([1, 2, 3, 4, 5])
w= np.sort(d, axis=-1,kind=None,order=None)
print(w)

```
